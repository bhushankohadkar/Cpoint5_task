```python
df_2=pd.read_excel(r"Inventory Dataset.xlsx")
df_2
```

```python
sorted_df = df_2.sort_values(by='Item Number', ascending=True).reset_index(drop=True)
sorted_df
```

```python
grouped_df = df_2.groupby('Item Number').agg({'Total Item Qty': 'sum'}).reset_index()

print("Grouped DataFrame by Item Number with Aggregation (e.g., Sum of Item Inventory):")
grouped_df
```

```python

df_2['Inventory Receive'] = pd.to_datetime(df_2['Inventory Receive'])
item_qty_sum = df_2.groupby('Item Number')['Total Item Qty'].sum().reset_index()

result = df_2.groupby(['Item Number', pd.Grouper(key='Inventory Receive', freq='7D')])['Total Item Qty'].sum().reset_index()

result_pivot = result.pivot(index='Item Number', columns='Inventory Receive', values='Total Item Qty').reset_index()

result_pivot.columns = ['Item Number'] + [col.strftime('%Y-%m-%d') for col in result_pivot.columns[1:]]

print("DataFrame with Item Number, Date, and Total Item Qty (Grouped by 7-day Intervals):")

result=result_pivot.fillna(0)
result
```

```python
item_qty_sum = df_2.groupby('Item Number')['Total Item Qty'].sum().reset_index()

result = df_2.groupby(['Item Number', pd.Grouper(key='Inventory Receive', freq='7D')])['Total Item Qty'].sum().reset_index()

result_pivot = result.pivot_table(index='Item Number', columns=pd.Grouper(key='Inventory Receive', freq='M'), values='Total Item Qty', aggfunc='sum').reset_index()

result_pivot_1 = result_pivot.fillna(0)

print("DataFrame with Item Number, Date, and Total Item Qty (Grouped by month):")
result_pivot_1
```

```python
column_mapping = {
    '2023-07-31 00:00:00': '2023-07-31',
    '2023-08-31 00:00:00': '2023-08-31',
    '2023-09-30 00:00:00': '2023-09-30',
    '2023-10-31 00:00:00': '2023-10-31',
    '2023-11-30 00:00:00': '2023-11-30',
    '2023-12-31 00:00:00': '2023-12-31',
  
}

result_pivot_1.rename(columns=column_mapping, inplace=True)

result_pivot_1
```

```python

column_names = [
    pd.Timestamp('2023-07-31 00:00:00'),
    pd.Timestamp('2023-08-31 00:00:00'),
    pd.Timestamp('2023-09-30 00:00:00'),
    pd.Timestamp('2023-10-31 00:00:00'),
    pd.Timestamp('2023-11-30 00:00:00'),
    pd.Timestamp('2023-12-31 00:00:00')
]

column_sums_df = pd.DataFrame(columns=['Inventory Receive', 'Total Item Qty'])

for col_name in column_names:
    col_sum = result_pivot_1[col_name].sum()
    column_sums_df = column_sums_df.append({'Inventory Receive': col_name, 'Total Item Qty': col_sum}, ignore_index=True)

column_sums_df
```

```python
plt.figure(figsize=(7, 4))
plt.bar(column_sums_df['Inventory Receive'], column_sums_df['Total Item Qty'], color='skyblue', edgecolor='black', width=10.0)
plt.xlabel('Inventory Receive', fontsize=12)
plt.ylabel('Total Item Qty', fontsize=12)
plt.title('Total Monthly Inventory Level', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
